## Self-similar solutions
Solutions of the steady, axisymmetric, non-relativistic ideal MHD equations assuming 
a spherically self-similar scaling were obtained by e.g. BP82
for the cold plasma outflow from the surface of a Keplerian disc. 
This solution for the field can be written in terms of variables $\chi$, $\xi(\chi)$, $\phi$, 
and $r_0$,  which are related to the cylindrical coordinates via

\begin{equation}\label{eq:r}
\mathbf{r} \equiv \left[r, \phi, z\right] = \left[r_0 \xi(\chi), \phi, r_0 \chi \right],%}
\end{equation}
where the adopted independent variables $(r_0, \chi)$ are a pair of spatial coordinates 
analogous to $(r, z)$.
The function $\xi(\chi)$ describes the shape of the field lines and, in the general case, is not 
\textit{a priori} known, but found as part of a self-consistent solution to the MHD equations. 
The flow velocity components are given by
\begin{equation}\label{eq:v}
\mathbf{v} = \left[\xi'(\chi)f(\chi), g(\chi), f(\chi) \right] \sqrt{\frac{G M}{r_0}},%}
\end{equation}
%
where a prime denotes differentiation with respect to $\chi$, and $G$ and $M$ 
are respectively the gravitational constant and the mass of the central black hole.  

In this self-similar model, the scaling of the speed $v$, magnetic field amplitude $B$, 
and gas density $\rho$ with the spherical radial coordinate $r$ is determined from 
the relation $B/\sqrt{\rho}  \propto r^{-1/2}$ and from the assumption that 
$r_0^2 \rho v_0$ is independent of $r_0$, from where $\rho_0 \propto r_0^{-3/2}$ 
and $B_0 \propto r_0^{-5/4}$. 
Other authors generalized this class of self-similar 
solutions by considering winds with a density scaling $\rho  \propto r^{-b}$, for 
which $B \propto r^{-(b+1)/2}$. 
Note that, in this context,  the BP82 solution corresponds to $b = 3/2$. The magnetic 
field and density at  arbitrary positions can be then written, in accordance with the 
self-similarity Ansatz, as ${\bf{ B} }= B_0(r_0) {\bf{b}}(\chi)$ and 
$\rho = \rho_0 (r_0) \varrho(\chi)$.
On the disc plane the rotational velocity, $v_{\phi}$, is Keplerian and scales as 
$v_{\phi} \propto r_0^{-1/2}$. The functions $\xi(\chi)$, $f(\chi)$ and $g(\chi)$ have 
to satisfy the flow MHD equations subject to the above scalings of $\rho$, $B$, and 
$v_{\phi}$ and boundary conditions. In particular, at the disc surface $\xi(0)=1$, 
$f(0)=0$ and $g(0)=1$.

Following BP82, we introduce the dimensionless expressions of the integrals of motion, in terms of which the solutions are defined:
$$
\kappa = k(1+\xi'_0)^{1/2}\frac{(GM/r_0)^{1/2}}{B_{0}}
$$
$$
\lambda = \frac{l}{(GMr_0)^{1/2}} \\
$$
$$
\epsilon = \frac{e}{(GM/r_0)}
$$
The parameters of the model are $\epsilon$, $\lambda$ and $\kappa$ and  $\xi'_0$. 
However, due to the regularity conditions that must be satisfied, these parameters 
are not independent. Combining the equations for $\lambda$ and $\epsilon$ gives 
$\epsilon~=~\lambda - \frac{3}{2}$.
The value of $\xi'_0\equiv \xi'(\chi=0)$ 
must be chosen to ensure the regularity of the solution at the Alfv\'en point. 
\footnote{
The Alfv\'en point is where the poloidal velocity of the fluid  is equal to the poloidal 
component $v_{\rm {pA}}$ of the Alfven velocity $v_{\rm {A}}$. The solutions are therefore parametrized only by two numbers, which can be chosen 
to be $\kappa$ and $\lambda$ (e.g. BP82). 

The Alfven speed,  $v_{\rm {A}}$,  is the characteristic velocity of the propagation of magnetic 
signals in an MHD fluid and is defined by: 
$$
v_{\rm {A}} = \frac{B^2}{\mu_0 \, \rho}, 
$$
where $\mu_0$ is the vacuum permeability. 
Another important characteristic quantity in magnetized fluids is the Alfv\'en Mach number at 
each position. The square of this quantity is expressed in the present model as 
$$
m(\chi) = \frac{v_{\rm {p}}^2}{v_{\rm {pA}}^2} = 
 \frac{4 \pi \rho \, v_{\rm {p}}^2}{B_{\rm {p}}^2} = \kappa \, f(\chi) \, \xi(\chi) \, J(\chi),
$$
where
$$
J(\chi) = \xi(\chi) - \chi \;\xi'(\chi)
$$
is the determinant of the Jacobian matrix of the transformation
$(r, z) \rightarrow (r_0, \chi)$ and $v_{\rm {pA}}$ is the 
poloidal component of the Alfv\'en velocity. 

The function $g(\chi)$ can be expressed in terms of the function $m$ and the specific
angular momentum, $\lambda$: 
$$
    g(\chi) = \frac{\xi^2(\chi) - \lambda \, m(\chi)}{\xi(\chi)\left[ 1 - m(\chi) \right]}. 
$$
From this expression we can see that the point corresponding to $m = 1$ is a singular 
point of the problem. In particular, to avoid unphysical solutions there, we must  have 
$\xi_{\rm {A}} = \xi(\chi_{\rm {A}})=\lambda^{1/2}$, where the subscript A refers to the 
Alfv\'en point. 

The model of EBS92 represents a simplified version of BP82 solution. 
The EBS92 solution corresponds to the case in which the solution asymptotically approaches 
$n =1$ as $\chi \rightarrow \infty$, where $n$ is the square of the Mach number for the fast 
magnetosonic mode for an arbitrary scaling of density  $\rho_0 \propto r_0^{-b}$ and magnetic 
field $B_0 \propto r_0^{-(b+1)/2}$.
The quantity $n$ is given by (e.g. BP82, BPS92): 
$$
n = \frac{4 \pi \rho \, v_{\rm {p}}^2}{B^2} = \frac{ \kappa \, \xi \,  f^3J  (1 + \xi'^2)}{T}. 
$$

While BP82 found their solutions by integrating a second-order differential equation, 
EBS92 impose {\it a priori} the functional form of the solution so that it will asymptotically 
tend to the BP82 solution.      
In their equation (3.19), EBS92 give an explicit form for the function $\xi(\chi)$:
$$
    \xi = \left( \frac{\chi}{c_2} +  1\right)^{1/2}, 
$$
where $c_2 = \frac{1}{2}\tan \vartheta_0$ was chosen to ensure that the field lines
make an initial angle $\vartheta_0$ with the disc plane, so that
$\cot \vartheta_0 = \xi'_0$ 
and the subscript $0$ means that the quantities are evaluated at the disc plane.
It can be demonstrated  \citep[e.g. BP82,][]{Heyvaerts96} that there is an upper 
limit for this angle: $\vartheta_0 < 60\degr$.

EBS92 found that to satisfy the condition that $n \rightarrow 1$ when $\chi \rightarrow \infty$, 
the parameters $\kappa$ and $\lambda$ must be related by:
$$
\kappa = 2 \left(\frac{3}{2 \lambda-3}\right)^{3/2}
$$
In addition, the asymptotic value of the function $f$ is given by 
$$
f_{\infty} =  \left(\frac{2 \lambda-3}{3}\right)^{1/2}.
$$
Thus, in this model the solutions depend on $\lambda$ and $\vartheta_0$.

Note that in the general case, in order to find the flow variables, we should have solved a 
second-order differential equation $ \xi''=\xi''(\chi, \xi, \xi', f(\chi))$, with $f(\chi)$ given 
implicitly by equation \ref{eq:BP82 Eq 2.12}. 
However, by using the EBS92 model we could evaluate $f'$, $m$ and $g'$ from an 
analytic estimate for $f$. The point of having used an analytical functional form for 
$f(\chi)$ has to do with the inclusion of the velocity field in the MC97 model. 
The opacity depends on the projection of the line of sight (LOS) component of the 
gradient of the LOS velocity through the quantity $Q$ (equation \ref{eq:Q-using-Lambda}), 
which involves the spatial derivatives of velocity components. 
Combining the EBS92 functional form for $\xi(\chi)$ and $\xi_{\rm {A}} = \lambda^{0.5}$ it 
is straightforward to obtain $\chi_{\rm {A}} = (\lambda-1) c_2$. 
In the general case this quantity must be found numerically as part of the solution. 
However, in the adopted framework, all related quantities at the Alfv\'en point are easily 
found (because $m_{\rm {A}} \equiv m(\chi_{\rm {A}}) = 1$
).  In particular, 
% 
\begin{equation}
f_{\rm {A}} = \frac{1}{\kappa \sqrt{\lambda} J_{\rm {A}}} = 
\frac{1}{\kappa\sqrt{\lambda}[\xi (\chi_{\rm {A}}) - \chi_{\rm {A}} \xi'(\chi_{\rm {A}})]}. 
\end{equation} 

The derivative of $f(\chi)$ at $\chi=0$, $f'(\chi=0)=f'_0$, is given by BP82 (their equation 2.23c), 
reproduced here: 
$$
f'_0=  \frac{(3\xi'^{2}_0-1)^{1/2} }{\kappa^2 \left[(\lambda-1)^2+(1+\xi'^{2}_0)\right]^{1/2}}\;.
$$
We thus adopt
$$
f(\chi)= f_\infty  \frac{ e^{k_1 \chi} - 1}{e^{k_1 \chi} + k_2} \;
$$
and look for $k_1$ and $k_2$ such that the conditions for $f_{\rm {A}}$ and $f'_0$ are satisfied. Once $f(\chi)$ is found, $m(\chi)$ and thus $g(\chi)$ are obtained. We then have the three wind velocity components expressed in analytical form. 


```python
import numpy as np 
%matplotlib inline 
import matplotlib.pyplot as plt 
from scipy.optimize import fsolve

def epsilon(chi, nu_0):
    '''
    function epsilon, =r/r0, defines the shape of streamlines and given by equation (32) of Chajet & Hall
    
    Arguments:
    chi -- the similarity variable/coordinate chi
    nu_0 -- angle the streamline makes with the disk plane at the footpoint
    '''
    c2 = 0.5 * np.tan(nu_0)
    return (np.sqrt(chi/c2 + 1.0))

def eps_prime(chi, nu_0):
    '''
    derivative of function epsilon
    
    Arguments:
    chi -- the similarity variable/coordinate chi
    nu_0 -- angle the streamline makes with the disk plane at the footpoint
    '''
    c2 = 0.5 * np.tan(nu_0)
    answer = 0.5 * (chi/c2 + 1.0) ** -0.5 * (1.0 / c2)
    #answer = 0.5 / np.sqrt(chi/c2 + 1.0) / c2
    return answer

def J(chi, nu_0):
    '''
    the determinant of the Jacobian of the similarity transformation(equation 27 of Chajet & Hall 2013)
    
    Arguments:
    chi -- the similarity variable/coordinate chi
    nu_0 -- angle the streamline makes with the disk plane at the footpoint
    '''
    answer = epsilon(chi, nu_0) - (chi * eps_prime(chi, nu_0))
    return (answer)

def kappa(lambd):
    '''
    the function kappa as defined in equation 22 of Chajet & Hall 2013
    
    Arguments:
    lambd -- the specific angular momentum
    '''
    answer = 2.0 * (3.0 / ((2.0 * lambd) - 3.0))**1.5
    return (answer)

def f_infty(lambd):
    '''
    the asymptotic value of the function f at infinity
    
    Arguments:
    lambd -- the specific angular momentum
    '''
    return (np.sqrt( ((2.0 * lambd) - 3.0) / 3.0) )

def fA(chiA, lambd, nu_0):
    '''
    the value of the function f at the Alfven point, chiA

    Arguments:
    chiA -- value of the similarity variable chi at the Alfven point
    lambd -- the specific angular momentum
    nu_0 -- angle the streamline makes with the disk plane at the footpoint
    '''
    k = kappa(lambd)
    answer = 1.0 / k / np.sqrt(lambd) /  J(chiA, nu_0)
    return (answer)

def fprime_0(lambd, nu_0):
    '''
    derivative of f at the streamline footpoint

    Arguments:
    lambd -- the specific angular momentum
    nu_0 -- angle the streamline makes with the disk plane at the footpoint
    '''
    eps_prime_0 = 1.0 / np.tan(nu_0)
    num = np.sqrt((3.0 * eps_prime_0 * eps_prime_0) - 1.0)
    k = kappa(lambd)
    denom = k * k * np.sqrt((lambd-1.0)**2 + (1.0-(eps_prime_0 * eps_prime_0)))
    return (num/denom)
    
def find_k1(lambd, chiA, nu_0):
    '''
    Find a numerical solution to find k1, in eq 37 of Chajet & Hall 2013
    
    Arguments:
    chiA -- value of the similarity variable chi at the Alfven point
    lambd -- the specific angular momentum
    nu_0 -- angle the streamline makes with the disk plane at the footpoint
    '''
    finf = f_infty(lambd)
    fprime0 = fprime_0(lambd, nu_0)
    f_alfven =fA(chiA, lambd, nu_0) 
    f_to_solve = lambda k: (finf * (np.exp(k * chiA) - 1.0) / (np.exp(k * chiA) - 1.0 + (finf * k / fprime0))) - f_alfven  
    ksol = fsolve(f_to_solve, 1.0)[0]
    print ("k1 is", ksol)
    return (ksol)

class MhdWindModel:
    def __init__(self, nu_0=20.0, lambd=10.0):
        self.nu_0 = np.radians(nu_0)
        self.lambd = lambd
        
        
    '''
    the next three functions are exactly the functions defined in Chajet & Hall and ESB92.
    They are define internally to the class because they require
    k1 and k2 to be known from numerical solution 
    '''    
    def f(self, chi, nu_0, lambd):
        answer = f_infty(lambd) * (np.exp(self.k1 * chi) - 1.0)
        answer /= (np.exp(self.k1 * chi) + self.k2)
        return (answer)
    
    def m(self, chi, nu_0, lambd):
        answer = kappa(lambd) * self.f(chi, nu_0, lambd) * epsilon(chi, nu_0) * J(chi, nu_0)
        return (answer)
    
    def g(self, chi, nu_0, lambd):
        num = (epsilon(chi, nu_0) ** 2) - (lambd * self.m(chi, nu_0, lambd))
        denom = epsilon(chi, nu_0) * (1.0 - self.m(chi, nu_0, lambd))
        return (num / denom)
        
    def setup_model(self, chi):
        
        self.chi = chi
 
        # find the Alfven point
        chi_alfven = (self.lambd - 1.0) * 0.5 * np.tan(self.nu_0)
        
        # find the asymptotic value of the function f
        finf = f_infty(self.lambd)
        
        # initial condition on the derivative of f
        fprime0 = fprime_0(self.lambd, self.nu_0)
        
        # find the constants in the form of the function f 
        self.k1 = find_k1(self.lambd, chi_alfven, self.nu_0)
        self.k2 = (finf * self.k1 / fprime0) - 1.0
        
        # get velocities
        self.vr = eps_prime(self.chi, self.nu_0) * self.f(self.chi, self.nu_0, self.lambd)
        self.vphi = self.g(self.chi, self.nu_0, self.lambd)
        self.vz = self.f(self.chi, self.nu_0, self.lambd)
           
```

The next cell calculates streamlines for a few different footpoints and plots them for two different launch angles $nu_0$

```python
chi = np.linspace(0,100,1000)

#nu_0 = np.radians(20)
angles = np.array([20,57])

# specific angular momentum?
lambd = 10.0

# footpoints for the streamlines
footpoints = np.linspace(0,100,30)

colors= ["C0", "C1"]
plt.figure(figsize=(11,5))
for i, nu_0 in enumerate(angles):
    mod = MhdWindModel(nu_0, lambd)
    mod.setup_model(chi)
    plt.subplot(1,2,i+1)
    
    for r0 in footpoints:
        r = epsilon(mod.chi, mod.nu_0) * r0
        z = mod.chi * r0
        plt.plot(r, z, c="C" + str(i))
        plt.plot(r, -z, c="C" + str(i))
        plt.plot(-r, z, c="C" + str(i))
        plt.plot(-r, -z, c="C" + str(i))
        
    plt.title(r"$\nu_0={}^\circ$".format(nu_0))
    plt.xlim(-100,100)
    plt.ylim(-100,100)
    if i == 0: plt.ylabel("$z$", fontsize=16)
    plt.xlabel("$r$", fontsize=16)
    #plt.savefig("streamlines_mhd.png") 
```
